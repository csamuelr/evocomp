```python
from utils.opposite_de import DifferentialEvolution
from utils.statistics import Statistics
from functions        import *
import matplotlib.pyplot as plt
from numpy import mean
from numpy import array
```

# Executando sem Opposite-based Learning

```python
all_best_solutions = []

print("Processando...")
for i in range(30):
    # print("Execução %d de 30" % (i+1))
    de = DifferentialEvolution(
        ng=100,
        np=200,
        cr=0.9, 
        f=0.8, 
        evfunc='Ackley', 
        algorithm='rand_1_bin',
        obl=False
    )
    de.evolve()
    all_best_solutions.append(de.get_all_best_solutions())
    
print("ok.")
```

```python
best_solutions = [] 
for i in range(100):
    best_solutions.append([])

for execution in all_best_solutions:
    for i, individual in enumerate(execution):            
        best_solutions[i].append(individual.get_fit())

mean_all_best_solutions = []
for pos in best_solutions:
    mean_all_best_solutions.append(mean(pos))
    
```

```python
plt.figure(figsize=(18,12))
plt.style.use('seaborn')
plt.title("Mean Without Opposition")
plt.plot(mean_all_best_solutions)
plt.xlabel('Generations')
plt.ylabel('Best Fitness')
plt.show()
```

# Executando com Opposite-based Learning

```python
all_opposite_best_solutions = []

print("Processando...")

for i in range(30):
    de = DifferentialEvolution(
        ng=100,
        np=200,
        cr=0.9, 
        f=0.8, 
        evfunc='Ackley', 
        algorithm='rand_1_bin',
        obl=True
    )
    de.evolve()
    all_opposite_best_solutions.append(de.get_all_best_solutions())

print("ok.")
```

```python
opposite_best_solutions = [] 
for i in range(100):
    opposite_best_solutions.append([])

for execution in all_opposite_best_solutions:
    for i, individual in enumerate(execution):            
        opposite_best_solutions[i].append(individual.get_fit())

mean_all_oppobest_solutions = []
for pos in opposite_best_solutions:
    mean_all_oppobest_solutions.append(mean(pos))
    
```

```python
plt.figure(figsize=(18,12))
plt.style.use('seaborn')
plt.title("Mean Without Opposition")
plt.plot(mean_all_oppobest_solutions, label='Mean Without Opposition')
plt.xlabel('Generations')
plt.ylabel('Best Fitness')
plt.show()
```

```python
plt.figure(figsize=(18,12))
plt.style.use('seaborn')
plt.title("Mean Without Opposition vs Mean With Opposition")
plt.plot(mean_all_best_solutions, label='Mean Without Opposition')
plt.plot(mean_all_oppobest_solutions, label='Mean Without Opposition')
plt.xlabel('Generations')
plt.ylabel('Best Fitness')
plt.show()
```
